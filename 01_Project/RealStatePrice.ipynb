```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("banglore_home_price.csv")
```

```python
# data.groupby("area_type")["area_type"].count()
data1 = data.drop(["area_type", "availability", "society"], axis=1)
data1["balcony"] = data1["balcony"].fillna(0)
data1.isna().sum()
data2 = data1.dropna(axis=0)
```

```python
data2["bhk"] = data2["size"].apply(lambda x: x.split(" ")[0])
# data2.head()
```

```python
data2["bath"] = data2["bath"].astype(int)
data2["balcony"] = data2["balcony"].astype(int)
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
data2[~data2["total_sqft"].apply(is_float)].head()
```

```python
def convertAvgToNum(x):
    token= x.split("-")
    if len(token) == 2:
        return (float(token[0]) + float(token[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data3 = data2.copy()
data3["total_sqft"] = data2["total_sqft"].apply(convertAvgToNum)
data3[data3["total_sqft"].apply(is_float)]
```
